## K-Means Clustering

Clustering is an unsupervised learning style that can provide some insights. Clustering algorithms attempt to group or co-locate samples that are similar or close to each other.

In this example, we will use the [K-means](https://en.wikipedia.org/wiki/K-means_clustering) algorithm to group three different species of iris, specifically iris Setosa, Versicolor, Virginica. K-means is a classic model that works well on a small data set. The [iris data set](https://en.wikipedia.org/wiki/Iris_flower_data_set) is available as part of [Sci-kit learn](https://scikit-learn.org/stable/), so it's nice place to start and demonstrates the basic idea.

![Iris Versicolor](images/iris_versicolor.png "Iris Versicolor")

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.target_names
```

```python
len(iris.data)
```

## Plot Sepal Width and Length

```python
X = iris.data[:, :2]
Y = iris.target

plt.scatter(X[:,0], X[:,1], c=Y)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

## Train Model with Three Clusters

```python
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

## Plot Predictions and Compare with Actual Values

```python
km_labels = km.labels_

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X[:, 0], X[:, 1], c=km_labels)
axes[1].scatter(X[:, 0], X[:, 1], c=Y)
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[0].set_title('Predicted')
axes[1].set_title('Actual')
```
